```python
China_1= China.ix[:93,:]

plt.plot(China_1['Std Dev in Stock Prices'], China_1['ROE (adjusted for R&D)'], 'bs')
plt.show()
```

```python
China_1.describe()
```

```python
mu = 0.073593  
sigma = 0.089079  
x = mu + sigma * China_1['Std Dev in Stock Prices']

num_bins = 30

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, normed=1)

# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
ax.plot(bins, y, '--')
ax.set_xlabel('ROE')
ax.set_ylabel('Probability density')
ax.set_title(r'Chinese Stock Return Distribution')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
